```python
import os
import librosa
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
home_path='''BDSMusicGenreClassification/'''
```

```python
def get_data(file_name, sample_rate):
#   print('++++++++File: {}'.format(file_name))
  file, sr = librosa.core.load(file_name, sr = sample_rate)
  gmm = librosa.feature.mfcc(file, sr=sr, n_mfcc=20)
  return pd.Series(np.hstack((np.mean(gmm, axis=1), np.std(gmm, axis=1))))
def visual_graph(file_name, sample_rate):
  data = get_data(file_name, sample_rate)
  plt.title('File: {}\nSample Rate: {}'.format(file_name,sample_rate))
  plt.plot(data)
  plt.show()
def print_data(file_name, sample_rate):
  data = get_data(file_name, sample_rate)
  print('File: {}\nSample Rate: {}'.format(file_name,sample_rate))
  print(data)
```

```python
G_num, M_num, SR_num = 5, 10, 3
SAMPLE_RATE = [8000, 11025, 16000, 22050, 32000, 37800, 44056, 44100, 47250, 48000, 50000, 50400, 88200, 96000, 176400, 192000, 352800, 2822400, 5644800, 11289600, 22579200]
for SR_num in range(10):
  genre_ls = os.listdir('''{}/genres'''.format(home_path))
  music_ls = os.listdir('''{}/genres/{}'''.format(home_path,genre_ls[G_num]))
  file_name = '''{}/genres/{}/{}'''.format(home_path,genre_ls[G_num],music_ls[M_num])
  visual_graph(file_name, SAMPLE_RATE[SR_num])
```

```python
df = pd.DataFrame([(this_genre, '''{}genres/{}/{}'''.format(home_path,this_genre,this_music))
                   for this_genre in genre_ls 
                   for this_music in os.listdir('''{}/genres/{}'''.format(home_path,this_genre))], columns=['genre','file_path'])
df.head()
```
